```r
library(readr)
data <- read_csv('regrex1.csv', show_col_types = FALSE)
```

```r
model <- lm(formula = y ~ x, data = data)
summary(model)
```

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = data$x, y = data$y), colour = 'blue') +
  geom_line(aes(x = data$x, y = predict(model, newdata = data)), colour = 'pink') +
  ggtitle('y vs x') +
  xlab('x') +
  ylab('y')
```
